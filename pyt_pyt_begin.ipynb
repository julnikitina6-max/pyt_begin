```python
def caesar_cipher(text):
    result = ""
    for char in text:
        if char.isalpha():                        # буква ли это
            shift_char = ord(char) + 3
            if char.isupper():                    # большая ли это буква
                result += chr(shift_char) if shift_char <= ord("Z") else chr(shift_char - 26)
            else:
                result += chr(shift_char) if shift_char <= ord("z") else chr(shift_char - 26)
        else:
            result += char
    return result

text = input("Введите фразу: ")
encrypted_text = caesar_cipher(text)
print("Зашифрованный текст: ", encrypted_text)
```

```python
def count_frequency (names_list):
   # Создаем пустой словарь для хранения результатов
    names_count = {}
         # Подсчитываем количество упоминаний каждого имени
    for name in names_list:
        if name in names_count:
            names_count[name] += 1
        else:
            names_count[name] = 1
          # Сортируем словарь по ключам (именам) в алфавитном порядке
    sorted_names_count = dict(sorted(names_count.items()))
    return sorted_names_count

```

```python
def flatten_nested_list (nested_list):
    flat_list = []
    for sublist in nested_list:
        if isinstance(sublist, list):
            flat_list.extend(flatten(sublist))  # рекурсивный вызов для обработки более глубоких уровней вложенности
        else:
            flat_list.append(sublist)
    return flat_list
```

```python
def process_commands(commands):
    results = []   # список для хранения результатов
    for index, command in enumerate(commands):
       results.append(index)  # добавляем индекс в результат
       if command == 'stop':
            break  # прекращаем выполнение
       elif command == 'skip':
            continue  # пропускаем команду
    return results
```

```python
def segment(num, scale):

    if not scale:
        raise ValueError("Список scale не должен быть пустым")

    if num < scale or num > scale[-1]:
        raise ValueError("Число num должно быть в диапазоне scale")

    # Находим минимальное и максимальное значение в scale
    min_value = scale
    max_value = scale[-1]

    return min_value, max_value
```

```python
def args_to_dict(*args, **kwargs):

    # Создаем словарь для хранения результатов
    result = {}

    # Добавляем позиционные аргументы в формате 'arg_N': значение
    for index, value in enumerate(args):
        result[f'arg_{index}'] = value

    # Добавляем именованные аргументы как есть
    result.update(kwargs)

    return result
```

```python
def safe_divide(num1, num2):
    try:
        # Попытка выполнить деление
        result = num1 / num2
        return result

    except ZeroDivisionError:
        # Обработка деления на ноль
        return "Ошибка: деление на ноль невозможно"

    except TypeError:
        # Обработка случаев, когда переданы не числа
        return "Ошибка: оба аргумента должны быть числами"

    except OverflowError:
        # Обработка переполнения (для очень больших чисел)
        return "Ошибка: результат слишком большой для представления"

    except Exception as e:
        # Обработка всех остальных исключений
        return f"Произошла непредвиденная ошибка: {str(e)}"
```

```python
import re

def validate_and_format_phones(phones):
    # Регулярное выражение для проверки номера
    pattern = re.compile(r'^(\+)?[0-9]{11}$|^(\+)?[0-9\s\-()]{11,20}$')

    valid_numbers = []

    for phone in phones:
        # Удаляем все допустимые разделители
        cleaned_phone = re.sub(r'[\s\-()]', '', phone)

        # Проверяем длину очищенного номера
        if len(cleaned_phone) != 11:
            continue

        # Проверяем формат с разделителями
        if not pattern.match(phone):
            continue

        # Проверяем, что разделители не стоят в начале
        if phone.startswith((' ', '-', '(')):
            continue

        # Если все проверки пройдены, добавляем в список валидных номеров
        valid_numbers.append(phone)

    return valid_numbers
```

```python
import re

def find_dates_in_text(text):
    # Регулярное выражение для поиска дат в формате YYYY-MM-DD
    r = r'\b\d{4}-\d{2}-\d{2}\b'

    # Находим все совпадения в тексте
    dates = re.findall(r, text)

    return dates
```

```python
import re

def extract_url_without_scheme(web_url: str) -> str:
    # Регулярное выражение для поиска схемы доступа
    scheme_pattern = r'^(https?|ftp):\/\/'

    # Регулярное выражение для удаления схемы доступа
    trim_pattern = r'^(https?|ftp):\/\/'

    # Поиск схемы доступа
    match = re.search(scheme_pattern, web_url)
    if match:
        scheme = match.group(0)
        print(f"Найденная схема доступа: {scheme}")

    # Удаление схемы доступа
    clean_url = re.sub(trim_pattern, '', web_url)

    return clean_url
```

```python
class Student:
    def __init__(self, name, student_id, major):
        self.name = name
        self.student_id = student_id
        self.major = major

    def get_info(self):
        return (f"Студент: {self.name}\n"
                f"ID: {self.student_id}\n"
                f"Специальность: {self.major}")

class GraduateStudent(Student):
    def __init__(self, name, student_id, major, thesis_topic, supervisor):
        super().__init__(name, student_id, major)
        self.thesis_topic = thesis_topic
        self.supervisor = supervisor

    def get_info(self):
        # Вызываем метод родительского класса
        base_info = super().get_info()
        # Добавляем дополнительную информацию
        return (f"{base_info}\n"
                f"Тема диссертации: {self.thesis_topic}\n"
                f"Научный руководитель: {self.supervisor}")
```

```python
import math

class Rectangle:
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        """Возвращает площадь прямоугольника"""
        return self.width * self.height

class Circle:
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        """Возвращает площадь круга"""
        return math.pi * (self.radius ** 2)

class Triangle:
    def __init__(self, a, b, c):
        self.a = a  # сторона a
        self.b = b  # сторона b
        self.c = c  # сторона c

    def area(self):
        """Возвращает площадь треугольника по формуле Герона"""
        # Полупериметр
        p = (self.a + self.b + self.c) / 2
        # Формула Герона
        return math.sqrt(p * (p - self.a) * (p - self.b) * (p - self.c))
```

```python
class Vector:
    def __init__(self, x: float, y: float, z: float):
        """Инициализация вектора с координатами x, y, z."""
        self.x = x
        self.y = y
        self.z = z

    def __str__(self):
        """Строковое представление вектора для удобного вывода."""
        return f"Vector({self.x}, {self.y}, {self.z})"

    def __add__(self, other):
        """Сложение двух векторов."""
        if isinstance(other, Vector):
            return Vector(
                self.x + other.x,
                self.y + other.y,
                self.z + other.z
            )
        return NotImplemented

    def __sub__(self, other):
        """Вычитание двух векторов."""
        if isinstance(other, Vector):
            return Vector(
                self.x - other.x,
                self.y - other.y,
                self.z - other.z
            )
        return NotImplemented

    def __mul__(self, scalar):
        """Умножение вектора на скаляр."""
        if isinstance(scalar, (int, float)):
            return Vector(
                self.x * scalar,
                self.y * scalar,
                self.z * scalar
            )
        return NotImplemented
```
